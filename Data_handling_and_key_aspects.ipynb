```python
print ( "Sample counts per class: \n " ,{ n :  v  for  n ,  v  in  zip ( breast_cancer.target_names ,  np.bincount (breast_cancer.target))})
```

```python
target_dict ={}
for i in range(2):
  target_dict[breast_cancer.target_names[i]] = np.bincount(breast_cancer.target)[i]
target_dict
```

```python
names = list(target_dict.keys())
values = list(target_dict.values())
plt.bar(0,values[0],tick_label=names[0])
plt.bar(1,values[1],tick_label=names[1])
plt.xticks(range(0,3),names)
plt.show()
```

```python
X = breast_cancer.data
y = breast_cancer.target
```

```python
plt.scatter(X[:,0],X[:,1],  c=y, cmap = "jet_r",alpha=0.5)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.title('Actual/Orignal Cluster')
plt.show()
```

```python
plt.scatter(breast_cancer.data[:,0],breast_cancer.data[:,7],  c=y, cmap = "jet_r",alpha=0.5)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Concave points')
plt.show()
```

```python
plt.scatter(breast_cancer.data[:,0],breast_cancer.data[:,9],  c=y, cmap = "jet_r",alpha=0.5)
plt.xlabel('Mean Radius')
plt.ylabel('Mean Fractal Dimension')
plt.show()
```

```python
plt.scatter(breast_cancer.data[:,10],breast_cancer.data[:,11],  c=y, cmap = "jet",alpha=0.5)
plt.xlabel('Radius Error')
plt.ylabel('Texture Error')
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='jet', edgecolor="k")
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel("Mean Radius")
ax.set_ylabel("Mean Texture")
ax.set_zlabel("Mean Perimeter")
ax.set_title('Actual/Orignal Cluster')
ax.dist=12
```
